```python
# 1. What is the role of the 'else' block in a try-except statement? Provide an example
# scenario where it would be useful


# In a try-except statement, the else block is executed if no exception occurs in the try block. The else block is optional.

# Here is an example :
try:
    num = int(input("Enter a number: "))
except ValueError:
    print("You did not enter a valid number!")
else:
    print("The number is:", num)

    
# explanation:
# In this example, the user is prompted to enter a number. The try block tries to convert the user input into an integer. If the user input is not a valid integer, the ValueError exception is raised and the except block is executed. If the user input is a valid integer, the else block is executed.

# The else block is useful in situations where you want to execute some code if no exception occurs in the try block. For example, in the above example, the else block prints the number that the user entered.
```

```python
# 2. Can a try-except block be nested inside another try-except block? Explain with an
# example.

# Yes, a try-except block can be nested inside another try-except block. This is called nested try-except blocks. Nested try-except blocks are used to handle exceptions that occur within other try blocks.

#example:
try:
    try:
        num1 = int(input("Enter the first number: "))
    except ValueError:
        print("You did not enter a valid number!")
    try:
        num2 = int(input("Enter the second number: "))
    except ValueError:
        print("You did not enter a valid number!")
    result = num1 / num2
except ZeroDivisionError:
    print("You cannot divide by zero!")

#Explanation:In this code, the first try block tries to convert the user input into integers. If the user input is not a valid integer, the ValueError exception is raised and the first except block is executed. The second try block tries to divide the two numbers. If the second number is zero, the ZeroDivisionError exception is raised and the second except block is executed.
```

```python

#3.How can you create a custom exception class in Python? Provide an example that demonstrates its usage.

class MyCustomError(Exception):
    def __init__(self, message="A custom error occurred"):
        self.message = message
        super().__init__(self.message)

# Usage example
def divide(a, b):
    if b == 0:
        raise MyCustomError("Division by zero is not allowed")
    return a / b

try:
    result = divide(10, 0)
except MyCustomError as e:
    print(f"Custom Exception Caught: {e}")
else:
    print(f"Result: {result}")
```

```python
##6 Explain the purpose of log levels in Python logging and provide examples of when each log level would be appropriate.

# In Python, the logging module provides a flexible and customizable way to record log messages in your application. Log levels in Python logging are used to categorize log messages based on their severity or importance. They allow you to filter and control which log messages get recorded and which ones are ignored. Python defines several standard log levels, each with its own purpose.

# here is example:

# 1.DEBUG:

# Purpose: The lowest log level used for debugging purposes. It provides detailed information about the application's internal state.
# Example: Debugging code, printing variable values, tracing the flow of a program.

import logging

logging.basicConfig(level=logging.DEBUG)
logging.debug("This is a debug message")

# 2.INFO:
# urpose: Provides information about the application's normal operation. It is often used for informational messages.
# Example: Recording successful startup, configuration changes, or other significant events.

import logging

logging.basicConfig(level=logging.INFO)
logging.info("Application started successfully")


# # 3.WARNING:
# # Purpose: Indicates a potential issue or a situation that may lead to problems but doesn't necessarily disrupt the application's operation.
# # Example: Warning about deprecated features or non-fatal errors.

import logging

logging.basicConfig(level=logging.WARNING)
logging.warning("Disk space is running low")

# 4.ERROR:
# Purpose: Indicates a critical error that prevents the application from functioning correctly.
# Example: Exception traceback messages, database connection failures.

import logging

logging.basicConfig(level=logging.ERROR)
try:
    result = 1 / 0
except ZeroDivisionError:
    logging.error("Error: Division by zero")


# 5.CRITICAL:

# Purpose: Denotes a severe error that may lead to application failure or data loss.
# Example: Critical system failures, security breaches, unhandled exceptions.

import logging

logging.basicConfig(level=logging.CRITICAL)
logging.critical("Security breach detected. Shutting down.")
```

```python
# What are log formatters in Python logging, and how can you customise the log
# message format using formatters?


# Log formatters in Python logging are responsible for converting a LogRecord object to an output string that can be interpreted by a human or external system. Formatters use a format string, which is a Python string that contains special placeholders for the different attributes of a LogRecord object.

import logging

# Create a formatter with your desired format
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')

# Create a console handler and set the formatter
console_handler = logging.StreamHandler()
console_handler.setFormatter(formatter)

# Create a logger, set its level, and add the console handler to it
logger = logging.getLogger('my_logger')
logger.setLevel(logging.DEBUG)
logger.addHandler(console_handler)

# Log messages with the configured logger
logger.debug('This is a debug message')
logger.info('This is an info message')
logger.warning('This is a warning message')




```

```python
# 8. How can you set up logging to capture log messages from multiple modules or
# classes in a Python application?


# To set up logging to capture log messages from multiple modules or classes in a Python application, you can follow these steps:

# 1.Create a Logger Instance for Each Module/Class:

# In each module or class where you want to use logging, create a logger instance by calling logging.getLogger() with a unique name.


# In module1.py
import logging

logger_module1 = logging.getLogger('module1')

# In module2.py
import logging

logger_module2 = logging.getLogger('module2')


# 2.Configure Logging at the Application Level:

# At the application level (typically in a central configuration file or script), configure the logging settings. This includes specifying log levels, handlers, formatters, and any other desired configurations.


import logging

# Configure root logger (global settings)
logging.basicConfig(level=logging.DEBUG, format='%(asctime)s - %(name)s - %(levelname)s - %(message)s')



# 3.Use Loggers in Modules/Classes:
# In each module or class, use the logger instance created in step 1 (logger_module1, logger_module2, etc.) to log messages at various log levels, such as debug(), info(), warning(), error(), and critical().


# In module1.py
logger_module1.debug('This is a debug message from module1')
logger_module1.info('This is an info message from module1')

# In module2.py
logger_module2.debug('This is a debug message from module2')
logger_module2.warning('This is a warning message from module2')

# 4.Configure Handlers and Formatters:
# You can configure different handlers and formatters for each logger if needed. For example, you may want to log messages from one module to a file and messages from another module to the console.

# 5.Control Logging Levels:
# Adjust the logging levels for each logger as needed. Different modules or classes may have different levels of verbosity. For example, you might set a lower level like DEBUG for debugging modules and a higher level like WARNING for critical parts of your application.

# By following these steps, you can set up logging to capture log messages from multiple modules or classes within your Python application. Each module or class can have its own logger instance with its own log settings, and the log messages will be appropriately categorized and formatted based on these settings.
```

```python
# 9.What is the difference between the logging and print statements in Python? When
# should you use logging over print statements in a real-world application?

# following are the difference between the logging and print statements in Python:
    
# 1.Purpose:

# Print Statements: print statements are primarily used for debugging and providing temporary output to the console during development. They are simple and easy to use for displaying information, variables, or debugging messages in your code.

# Logging: The logging module is designed for more robust and structured handling of messages and events in your application. It is used for generating log files and managing log levels, making it suitable for production and long-term monitoring.

# 2.Output Destination:

# Print Statements: print statements send output directly to the console (standard output). They are visible in the terminal or command prompt where you run your script.

# Logging: Logging allows you to direct output to various destinations, such as log files, the console, or external services. This flexibility is valuable for different use cases, including debugging, monitoring, and auditing.

# 3.Log Levels:

# Print Statements: print statements do not support log levels by default. Every message is treated the same way and is always displayed.

# Logging: Logging supports different log levels (DEBUG, INFO, WARNING, ERROR, CRITICAL), allowing you to categorize and filter messages based on their importance. This helps in controlling the verbosity of log output.

# 4.Configurability:

# Print Statements: print statements are simple and do not offer advanced configuration options. You can't easily change their behavior or format across different parts of your code.

# Logging: The logging module provides extensive configuration options. You can customize the log format, specify log levels for different parts of your application, and control where log messages are stored (e.g., files, external services). This configurability is essential for managing complex real-world applications.

# 5.Long-Term Maintenance:

# Print Statements: Using print statements for debugging is convenient during development but is not suitable for long-term maintenance or monitoring in production environments.

# Logging: Logging is designed for long-term use. You can leave logging statements in your code and change their behavior or destinations without modifying the code itself. This is valuable for diagnosing issues and monitoring application behavior in production.

# In a real-world application, you should use logging over print statements when:

# 1.You need to maintain a history of events and troubleshoot issues in production environments.
# 2.You want to categorize and prioritize messages based on their importance.
# 3.You need to control where log messages are stored (e.g., log files) and potentially send them to external services.
# 4.You are building a robust and maintainable software system.
```

```python
# 10. Write a Python program that logs a message to a file named "app.log" with the
# following requirements:
# ● The log message should be "Hello, World!"
# ● The log level should be set to "INFO."
# ● The log file should append new log entries without overwriting previous ones.



import logging

# Configure logging to write to app.log file
logging.basicConfig(filename='app.log', level=logging.INFO, filemode='a', format='%(asctime)s - %(levelname)s - %(message)s')

# Log the message
logging.info('Hello, World!')

# Add a success message
print('Log entry added successfully.')


# explanation:1.We import the logging module.

# 2.We configure logging using basicConfig to set up the log file, log level, and formatting.
# filename='app.log': Specifies the name of the log file.
# level=logging.INFO: Sets the log level to INFO.
# filemode='a': Specifies that new log entries should be appended to the existing log file without overwriting it.
# format='%(asctime)s - %(levelname)s - %(message)s': Defines the format of the log entries, including the timestamp, log level, and message.

# 3.We use logging.info('Hello, World!') to log the message "Hello, World!" with an INFO log level.
```
